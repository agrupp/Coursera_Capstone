```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as soup
import requests
```

Getting and parsing html from url provided

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
doc = requests.get(url)
html = doc.text
tree = soup(html,'html.parser')
table = tree.table.find_all('td')
```

Extracting table values

```python
iter = 0
Pos = []
Bor = []
Nei = []

for i in table:
    data = i.text.rstrip()
    iter = iter + 1
    if iter == 1: 
        Pos.append(data)
    if iter == 2: 
        Bor.append(data)
    if iter == 3: 
        Nei.append(data)
        iter = 0
```

Building dataframe from extracted values

```python
df = pd.DataFrame()
df['Postal Code'] = Pos
df['Borough'] = Bor
df['Neighborhood'] = Nei

print(df.head(10))
```
